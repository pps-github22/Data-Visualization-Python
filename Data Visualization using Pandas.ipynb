# Case study 1

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
chrun = pd.read_csv("Customer_Churn.csv",nrows=1000)
plt.bar(chrun["Contract"].value_counts().keys().tolist(),chrun["Contract"].value_counts().tolist(),color ="orange")
plt.xlabel("Contract Type of customer")
plt.ylabel("Count")
plt.title("Distribution of Contract")
```

```python
plt.hist(chrun["MonthlyCharges"],color = "forestgreen")
plt.xlabel("Monthly charges Incurred")
plt.ylabel("Count")
plt.title("Distribution of Monthly Charges")
```

```python
import seaborn as sb
sb.scatterplot(x = chrun["tenure"],y = chrun["TotalCharges"], color = "indigo")
plt.xlabel("Tenure of the customer")
plt.ylabel("Total chargesIncurred")
plt.title("Total charges vs Tenure")
```

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
chrun = pd.read_csv("Customer_Churn.csv")
sb.boxplot(x = chrun["PaymentMethod"], y = chrun["MonthlyCharges"], color = "olive")
plt.xticks(rotation=45)
plt.xlabel("Paymentmethod of Customers")
plt.ylabel("MonthlyCharges of Customers")
plt.title("Paymentmethod VS MonthlyCharges")
```
